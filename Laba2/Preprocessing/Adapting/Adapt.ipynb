# Импорт

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from imblearn.over_sampling import SMOTE
```

# Загружаем DataSet

```python
models = pd.read_csv('../../../Data/DataLaba2/neo_task_moded_norm.csv')
models
```

```python
X, y = models.drop('hazardous', axis=1), models['hazardous']
```

```python
X_balanced, y_balanced = SMOTE().fit_resample(X, y)
```

```python
X_balanced['hazardous'] = y_balanced
```

```python
X_balanced
```

```python

X_balanced.to_csv('../../../Data/DataLaba2/neo_task_smoted.csv', index=False)
```
