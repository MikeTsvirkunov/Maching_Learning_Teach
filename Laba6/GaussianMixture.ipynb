```python
from sklearn.mixture import GaussianMixture
import matplotlib.gridspec as gridspec
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import silhouette_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestNeighbors

from sklearn.model_selection import GridSearchCV

from sklearn.metrics import rand_score, adjusted_rand_score, fowlkes_mallows_score

from joblib import dump
```

# Data1

```python
data1 = pd.read_csv('../Data/DataLaba6/data1.csv')
x1 = (data1.drop('class', axis=1)).to_numpy()
y1 = data1['class'].to_numpy()
```

```python
means = []
covariances = []
for i in range(1, 11):
    km = GaussianMixture(n_components=i).fit(x1)
    means.append(km.means_.prod(axis=1).sum(axis=0))
    covariances.append(km.covariances_)
plt.plot(np.array(means), c='r')
```

```python
ss = []
for i in range(2, 11):
    ss.append(silhouette_score(x1, GaussianMixture(n_components=i).fit_predict(x1)))
plt.plot(ss)
plt.xticks(range(0, 9), range(2, 11))
```

```python
gausmix1 = GaussianMixture(n_components=2).fit(x1)
fig = plt.figure(tight_layout=True)
fig.set_figheight(5)
fig.set_figwidth(15)
gs = gridspec.GridSpec(1, 2)
ax = fig.add_subplot(gs[0, 0])
ax.scatter(x1[:, 0], x1[:, 1], c=gausmix1.predict(x1))
ax = fig.add_subplot(gs[0, 1])
ax.scatter(x1[:,0], x1[:,1], c=y1)
```

```python
rand_score(y1, gausmix1.predict(x1)), adjusted_rand_score(y1, gausmix1.predict(x1)), fowlkes_mallows_score(y1, gausmix1.predict(x1))
```
