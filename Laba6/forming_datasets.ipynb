```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas as pd
```

```python
data1 = make_blobs(n_samples=200, n_features=2, centers=3)
```

```python
plt.scatter(data1[0][:,0], data1[0][:,1], c=data1[1])
```

```python
df1 = pd.DataFrame(data1[0])
df1.columns = ['param1', 'param2']
df1['class'] = data1[1]
df1.to_csv('../Data/DataLaba6/data1.csv', index=0)
```
