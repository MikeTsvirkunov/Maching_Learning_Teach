```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_classification
import pandas as pd
```

```python
data1 = make_classification(n_samples=200, 
                            n_features=2,
                            n_informative=2, 
                            n_redundant=0, 
                            n_repeated=0, 
                            n_classes=2,
                            n_clusters_per_class=1, 
                            weights=np.array([0.5, 0.4]), 
#                             # flip_y=0.01, 
#                             # class_sep=1.0, 
#                             # hypercube=True, 
#                             # shift=0.0, 
#                             # scale=1.0, 
#                             # shuffle=True, 
#                             # random_state=None
                            )
plt.scatter(data1[0][:,0], data1[0][:,1], c=data1[1])
```

```python
data1 = make_blobs(n_samples=200, n_features=2, centers=3, cluster_std=2.8)
plt.scatter(data1[0][:,0], data1[0][:,1], c=data1[1])
```

```python
df1 = pd.DataFrame(data1[0])
df1.columns = ['param1', 'param2']
df1['class'] = data1[1]
df1.to_csv('../Data/DataLaba6/data1.csv', index=0)
```
