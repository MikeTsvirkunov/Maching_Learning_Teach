```python
from sklearn_som.som import SOM
import matplotlib.gridspec as gridspec
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import silhouette_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestNeighbors

from sklearn.model_selection import GridSearchCV

from sklearn.metrics import rand_score, adjusted_rand_score

from joblib import dump
```

# Data1

```python
data1 = pd.read_csv('../Data/DataLaba6/data1.csv')
x1 = (data1.drop('class', axis=1)).to_numpy()
y1 = data1['class'].to_numpy()
```

```python
ss = [-10, 1, 1]
for m, n in [(m, n) for m in range(2, 15) for n in range(1, 5)]:
    s = SOM(m=m, n=n, dim=2)
    s.fit(x1)
    p = s.predict(x1)
    sst = [silhouette_score(x1, p), m, n] if len(np.unique(p)) > 1 else ss
    ss = max(ss, sst, key=lambda a: a[0])
ss
```

```python
som1 = SOM(m=3, n=1, dim=2)
som1.fit(x1)
```

```python
fig = plt.figure(tight_layout=True)
fig.set_figheight(5)
fig.set_figwidth(15)
gs = gridspec.GridSpec(1, 2)
ax = fig.add_subplot(gs[0, 0])
ax.scatter(x1[:, 0], x1[:, 1], c=som1.predict(x1))
ax = fig.add_subplot(gs[0, 1])
ax.scatter(x1[:, 0], x1[:, 1], c=y1)
```
