```python
from scipy.cluster import hierarchy
import matplotlib.gridspec as gridspec
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import silhouette_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestNeighbors

from sklearn.model_selection import GridSearchCV

from sklearn.metrics import rand_score, adjusted_rand_score

from joblib import dump
```

# Data1

```python
data1 = pd.read_csv('../Data/DataLaba6/data1.csv')
x1 = (data1.drop('class', axis=1)).to_numpy()
y1 = data1['class'].to_numpy()
```

```python
linked = hierarchy.linkage(x1, method='ward')
```

```python
ss = []
for i in np.arange(2, 100):
    c = hierarchy.fcluster(linked, t=i, criterion='distance')
    nc = np.unique(c).shape[0]
    # print(i, nc)
    ss.append(silhouette_score(x1, c) if nc > 1 and nc < x1.shape[0]  else -1)
plt.plot(ss)
plt.axvline(15, color='red', linestyle='-')
# ss
# plt.xticks(range(0, 18), np.arange(0, 9, 0.5))
```

```python
plt.figure(figsize=(8, 6))
hierarchy.dendrogram(linked)
plt.axhline(3.5, color='red', linestyle='--')
plt.axhline(6, color='red', linestyle=':')
plt.axhline(20, color='red', linestyle='-')
```

```python
p1 = hierarchy.fcluster(hierarchy.linkage(x1, method="ward"), t=20, criterion='distance')
```

```python
fig = plt.figure(tight_layout=True)
fig.set_figheight(5)
fig.set_figwidth(15)
gs = gridspec.GridSpec(1, 2)
ax = fig.add_subplot(gs[0, 0])
ax.scatter(x1[:, 0], x1[:, 1], c=p1)
ax = fig.add_subplot(gs[0, 1])
ax.scatter(x1[:,0], x1[:,1], c=y1)
```
