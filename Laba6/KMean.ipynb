```python
from sklearn.cluster import KMeans
import matplotlib.gridspec as gridspec
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import silhouette_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import silhouette_score

from sklearn.model_selection import GridSearchCV

from sklearn.metrics import rand_score, adjusted_rand_score, fowlkes_mallows_score

from joblib import dump
```

# Data1

```python
data1 = pd.read_csv('../Data/DataLaba6/data1.csv')
x1 = (data1.drop('class', axis=1)).to_numpy()
y1 = data1['class'].to_numpy()
```

```python
inertias = []
for i in range(2, 11):
    km = KMeans(i, n_init='auto').fit(x1)
    inertias.append(km.inertia_)
plt.plot(inertias)
plt.xticks(range(0, 9), range(2, 11))

```

```python
ss = []
for i in range(2, 11):
    ss.append(silhouette_score(x1, KMeans(i, n_init='auto').fit(x1).labels_))
plt.plot(ss)
plt.xticks(range(0, 9), range(2, 11))
```

```python
kmeans1 = KMeans(3, n_init='auto').fit(x1)
```

```python
fig = plt.figure(tight_layout=True)
fig.set_figheight(5)
fig.set_figwidth(15)
gs = gridspec.GridSpec(1, 2)
ax = fig.add_subplot(gs[0, 0])
ax.scatter(x1[:,0], x1[:,1], c=kmeans1.labels_)
ax.scatter(kmeans1.cluster_centers_[:, 0],
           kmeans1.cluster_centers_[:, 1], s=100, c='red')
ax = fig.add_subplot(gs[0, 1])
ax.scatter(x1[:,0], x1[:,1], c=y1)
```

```python
rand_score(y1, kmeans1.labels_), adjusted_rand_score(y1, kmeans1.labels_), fowlkes_mallows_score(y1, kmeans1.labels_)
```
