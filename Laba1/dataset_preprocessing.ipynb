## Нивелирование анамалий

## Нивелирование анамалий
- Заполнение пропущенных значений (рекомедуется логика заполнения пропусков на основе типа данных, которая использовалась в РГР по Практикуму)
- Удаление дубликатов

```python
df = df.drop_duplicates()
df = df.apply(lambda a: a.fillna(a.mean() if a.dtypes ==
              np.float64 else a.median() if a.dtypes == np.int64 else a.mode()[0]))
df
```

```python
df.isnull().sum()
```

# Преобразование даты

```python
d = df['date'].str.split(expand=True)[0].str.split(expand=True, pat='-').assign(time=df['date'].str.split(expand=True)[1].map(lambda a: int(a.split(':')[0])*60 + int(a.split(':')[1]))).apply(lambda a: a.map(int))
d.columns = ['day', 'month', 'year', 'time']
df = pd.concat([df.drop('date', axis=1), d], axis=1, join="inner")
df
```

```python
d = df_removed_nan['date'].str.split(expand=True)[0].str.split(expand=True, pat='-').assign(time=df_removed_nan['date'].str.split(expand=True)[1].map(lambda a: int(a.split(':')[0])*60 + int(a.split(':')[1]))).apply(lambda a: a.map(int))
d.columns = ['day', 'month', 'year', 'time']
df_removed_nan = pd.concat([df_removed_nan.drop('date', axis=1), d], axis=1, join="inner")
df_removed_nan
```

## Преобразование категориальных признаков в числовые (используйте one-hot кодирование или map; используйте знания с Практикума)

```python
names = dict(df.loc[:, list(filter(lambda a: df[a].dtypes == 'object', df.columns))].apply(lambda a: a.unique()))
names
```

```python
def replacer(a):
    return a.map(lambda k: np.where(names[a.name] == k)[0][0]) if a.dtypes == 'object' else a
```
