```python
import numpy as np
import pandas as pd
from Classes.PercepTeacher import PerceptronTeacher
from Functions.function_of_activation import sigmoid, dsigmoid
from Functions.errorFunctions import MSE
from Classes.Layer import Layer
from Classes.LayerGen import LayerGenerator
from random import random
from Classes.Perceptron import Perceptron
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from Classes.LossFunction import LossFunction
from sklearn.metrics import classification_report
from Functions.errorFunctions import R2
from Classes.perec import WeightsSearch
np.random.seed(1)
```

```python
data = pd.read_csv('../Data/DataLaba1/energy_task_moded_removed_nan.csv')
sl = data.shape[0]
scale = MinMaxScaler().fit_transform(data)
X_train, X_test, y_train, y_test = train_test_split(
    scale[:, :-2], scale[:, -1], test_size=0.25)
```

```python
layers_dims = [28, 64, 128, 64, 32, 16, 1]
```

```python
neuron = WeightsSearch(layers_dims,  learning_rate=0.00001, num_iterations = 1000)
```

```python
train = neuron.search(X_train, y_train)
```

```python
pred_test = neuron.predict_regr(X_test, train[0])
```

```python
layers = []
for i in range(0, len(train[0]), 2):
    w = train[0][list(train[0].keys())[i]]
    d = train[0][list(train[0].keys())[i+1]]
    print(w.shape, d.shape)
    layers.append(Layer(w, d, sigmoid, dsigmoid))
```

```python
a = Perceptron(layers)
res = np.array([a.predict(i)[0] for i in X_train[:200]])
# print(MSE((res), y_train[:200]))
for p, y in zip(res, y_train[:200]):
    print(f'p:{p},\t y:{y},\t d:{abs(p-y)}')
```
