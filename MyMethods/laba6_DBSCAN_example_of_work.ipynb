```python
from Classes.DBSCANTrainer import DBSCANTrainer
from Classes.DBSCAN import DBSCAN
from Functions.functions_of_priority import max_count_class
import numpy as np
import pandas as pd
from random import choice
from Functions.distance_functions import euclid_distance
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from sklearn.metrics import classification_report, confusion_matrix
```

```python
data1 = pd.read_csv('../Data/DataLaba6/data1.csv')
x1 = (data1.drop('class', axis=1)).to_numpy()
y1 = data1['class'].to_numpy()
```

```python
z = DBSCANTrainer(0.8, euclid_distance, 1).train(x1)
```

```python
x = []
y = []
for i in z:
    for k in z[i]:
        y.append(i)
        x.append(k)
x = np.array(x)
```

```python
fig = plt.figure(tight_layout=True)
fig.set_figheight(5)
fig.set_figwidth(15)
gs = gridspec.GridSpec(1, 2)
ax = fig.add_subplot(gs[0, 0])
ax.scatter(x[:, 0], x[:, 1], c=y)
ax = fig.add_subplot(gs[0, 1])
ax.scatter(x1[:,0], x1[:,1], c=y1)
```
